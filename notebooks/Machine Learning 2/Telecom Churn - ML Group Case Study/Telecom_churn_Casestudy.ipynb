### Filter high-value customers
We need to predict churn only for the high-value customers. Define high-value customers as follows: Those who have recharged with an amount more than or equal to X, where X is the 70th percentile of the average recharge amount in the first two months (the good phase).

````python
# Getting total amount of recharge done each month for data
for month in [6, 7, 8, 9]:
    df_telecom_churn['total_rech_data_amt_' + str(month)] = df_telecom_churn['av_rech_amt_data_' + str(month)] * df_telecom_churn['total_rech_data_' + str(month)]
    df_telecom_churn.drop(['av_rech_amt_data_' + str(month), 'total_rech_data_' + str(month)], axis = 1, inplace = True)
````

````python
# Getting avarage recharge done for month 6 and 7 i.e. the good phase for both data and calling
df_telecom_churn['telecom_avg_rech_6_7'] = (df_telecom_churn['total_rech_amt_6'].fillna(0) + df_telecom_churn['total_rech_amt_7'].fillna(0) + df_telecom_churn['total_rech_data_amt_6'].fillna(0) + df_telecom_churn['total_rech_data_amt_7'].fillna(0))/2

# Finding the 70th percentile for high valued customer
percentile_70 = np.percentile(df_telecom_churn['telecom_avg_rech_6_7'], 70)
percentile_70
````

````python
# Filtering customers having average recharge greater than 70th percentile
df_telecom_hv_churn = df_telecom_churn[df_telecom_churn['telecom_avg_rech_6_7'] > percentile_70]
df_telecom_hv_churn.shape
````

````python
# create plotting functions
def data_type(variable):
    if variable.dtype == np.int64 or variable.dtype == np.float64:
        return 'numerical'
    elif variable.dtype == 'category':
        return 'categorical'
    
def univariate(variable, stats=True):
    
    if data_type(variable) == 'numerical':
        sns.distplot(variable)
        if stats == True:
            print(variable.describe())
    
    elif data_type(variable) == 'categorical':
        sns.countplot(variable)
        if stats == True:
            print(variable.value_counts())
            
    else:
        print("Invalid variable passed: either pass a numeric variable or a categorical vairable.")
````

### Modelling

#### Preparing training and testing data

````python
# we will create a new dataset for model building
df = telecom_hv_cust[:].copy()

# lets create X & y dataset for model building
# X with "churn" column and y with only have "churn" column

X = df.drop(['churn'], axis=1)
y = df['churn']

df.drop('churn', axis=1, inplace=True)
````

````python
# apply scaling on the dataset

from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler

scaler = preprocessing.StandardScaler().fit(X)
X = scaler.transform(X)

# Splitting data into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, train_size=0.7, random_state=1)
````

````python
# applying PCA to select the top 20 columns

from sklearn.decomposition import PCA
pca = PCA(random_state=42)
````

````python
pca.fit(X_train)
````

````python
pca.components_
````

````python
pca.explained_variance_ratio_
````

````python
plt.bar(range(1,len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_)
````

````python
var_cumu = np.cumsum(pca.explained_variance_ratio_)
plt.plot(range(1,len(var_cumu)+1), var_cumu)
````

#### As we can see that the cumulative value increases rapidly upto 0.9 with the number of predictors around 60 and then it takes all the remaining 27 values to get to 1.0. So, we will take 0.9 as our cut-off cumulative value


````python
pca = PCA(0.90, random_state=42)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
X_train_pca.shape
````

#### As we can see that now we have only 45 predictors left now

### Hyperparameter Tuning

### Tuning max_depth

Let's try to find the optimum values for ```max_depth``` and understand how the value of max_depth impacts the overall accuracy of the ensemble.


````python
# GridSearchCV to find optimal n_estimators
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV


# specify number of folds for k-fold CV
n_folds = 5

# parameters to build the model on
parameters = {'max_depth': range(2, 20, 5)}

# instantiate the model
rf = RandomForestClassifier()

# fit tree on training data
rf = GridSearchCV(rf, parameters, 
                    cv=n_folds, 
                   scoring="accuracy",
                 return_train_score=True)
rf.fit(X_train_pca, y_train)
````

````python
# scores of GridSearch CV
scores = rf.cv_results_
pd.DataFrame(scores).head()
````

````python
# plotting accuracies with max_depth
plt.figure()
plt.plot(scores["param_max_depth"], 
         scores["mean_train_score"], 
         label="training accuracy")
plt.plot(scores["param_max_depth"], 
         scores["mean_test_score"], 
         label="test accuracy")
plt.xlabel("max_depth")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
````

You can see that as we increase the value of max_depth, both train and test scores are stable till a point, but after that train score starts to increase and test score seems to be constant.

Thus, controlling the depth of the constituent trees will help reduce overfitting in the forest.

### Tuning n_estimators

### Tuning min_samples_leaf

The hyperparameter **min_samples_leaf** is the minimum number of samples required to be at a leaf node:
- If int, then consider min_samples_leaf as the minimum number.
- If float, then min_samples_leaf is a percentage and ceil(min_samples_leaf * n_samples) are the minimum number of samples for each node.

Let's now check the optimum value for min samples leaf in our case.

````python
# GridSearchCV to find optimal min_samples_leaf
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV


# specify number of folds for k-fold CV
n_folds = 5

# parameters to build the model on
parameters = {'min_samples_leaf': range(100, 400, 50)}

# instantiate the model
rf = RandomForestClassifier()


# fit tree on training data
rf = GridSearchCV(rf, parameters, 
                    cv=n_folds, 
                   scoring="accuracy",
                 return_train_score=True)
rf.fit(X_train_pca, y_train)
````

````python
# scores of GridSearch CV
scores = rf.cv_results_
pd.DataFrame(scores).head()
````

````python
# plotting accuracies with min_samples_leaf
plt.figure()
plt.plot(scores["param_min_samples_leaf"], 
         scores["mean_train_score"], 
         label="training accuracy")
plt.plot(scores["param_min_samples_leaf"], 
         scores["mean_test_score"], 
         label="test accuracy")
plt.xlabel("min_samples_leaf")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
````

You can see that the model starts of overfit as you decrease the value of min_samples_leaf.
