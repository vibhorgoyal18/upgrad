```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#set it to None to remove SettingWithCopyWarning
pd.options.mode.chained_assignment = None 
#set it to convert scientific noations to float
pd.options.display.float_format = '{:.4f}'.format 
```

```python
# Loading data into dataframes
file_location = "C:/Users/chebrolb/loan.csv"
loan_df = pd.read_csv(file_location, sep=',', dtype='unicode')
loan_df_dummy = loan_df.copy()
```

```python
loan_df.head()
print(loan_df.shape)
print(loan_df_dummy.shape)
print(loan_df_dummy.info())
```

```python
# Data Cleaning

# Remove columns giving no value
not_required_columns_in_loandf = ["id","member_id","url","zip_code","desc"]
for i in range(len(not_required_columns_in_loandf)):
    if not_required_columns_in_loandf[i] in loan_df_dummy.columns:
        loan_df_dummy.drop(not_required_columns_in_loandf, axis =1, inplace=True)
    else:
        print(not_required_columns_in_loandf[i] + " is not present")
print(loan_df_dummy.shape)

# Convert quantitative columns to numeric datatype
numeric_columns = ['loan_amnt','funded_amnt','funded_amnt_inv','installment','annual_inc','dti', 'last_pymnt_amnt', 'collection_recovery_fee',
                  'recoveries', 'total_rec_late_fee', 'total_rec_int', 'total_rec_prncp', 'total_pymnt_inv', 'total_pymnt', 'out_prncp_inv',
                  'out_prncp', 'total_acc', 'revol_bal','delinq_2yrs']
loan_df_dummy[numeric_columns] = loan_df_dummy[numeric_columns].apply(pd.to_numeric)


# Employment Term : Remove 'n/a' value with 'self-employed'.
loan_df_dummy.emp_length.replace(['n/a'],'Self-Employed',inplace=True)

# Interest Rate : Replace Blanks with 0.
loan_df_dummy["int_rate"].fillna(0, inplace = True)

# Employee Title : Replace Blanks with "No Title Provided".
loan_df_dummy["emp_title"].fillna("No Title Provided", inplace = True)

# Assert column 'loan_amnt' has no null values because if there are any null values in column 'loan_amnt' 
# we can directly remove those rows as they don't give any value from them
assert(~loan_df_dummy['loan_amnt'].isnull().any())

#Ensure to not to have null values in column 'funded_amnt', if there are any need to impute those missing values
assert(~loan_df_dummy['funded_amnt'].isnull().any())

#Ensure that all values in column 'term' is expressed in months 
assert(loan_df_dummy['term'].str.contains('months').all())

#Ensure that all values in column 'annual income' is having no null values 
assert(~loan_df_dummy['annual_inc'].isnull().any())

# Remove columns having single unique value as they will not give value for our analysis
def find_columns_with_single_unique_value(dataFrame):
    columns_with_single_unique_value = []
    for column in list(dataFrame.columns):
        if dataFrame[column].unique().size < 2:
            columns_with_single_unique_value.append(column)
    return columns_with_single_unique_value
columns_with_single_unique_value = find_columns_with_single_unique_value(loan_df_dummy)
print(columns_with_single_unique_value)
loan_df_dummy.drop(columns_with_single_unique_value, axis = 1, inplace = True)
print(loan_df_dummy.shape)

# List of Columns & NA counts where NA values are more than 30%
NA_columns = loan_df_dummy.isnull().sum()
NA_columns = NA_columns[NA_columns.values >=(0.3*len(loan_df_dummy))].index
loan_df_dummy.drop(labels = NA_columns, axis = 1, inplace = True)
print(loan_df_dummy.shape)
# Output : mths_since_last_delinq, mths_since_last_record, next_pymnt_d

# Drop duplicates
loan_df_dummy.drop_duplicates(inplace= True)
print(loan_df_dummy.shape)

# Loan Status : Drop records where values are less than 1.5%
loan_status_delete_records = (loan_df_dummy.loan_status.value_counts()*100)/len(loan_df_dummy)
loan_status_delete_records = loan_status_delete_records[(loan_status_delete_records < 1.5)]
loan_df_dummy.drop(loan_df_dummy[loan_df_dummy.loan_status.isin(loan_status_delete_records.index)].index, inplace=True)
print(loan_df_dummy.loan_status.unique())

# Remove rows where home_ownership'=='OTHER', 'NONE', 'ANY'
unwanted_entries = ['OTHER', 'NONE', 'ANY']
loan_df_dummy.drop(loan_df_dummy[loan_df_dummy['home_ownership'].isin(unwanted_entries)].index,inplace=True)
```

```python
# Data Manipulation

# Loan amount to Annual Income ratio
loan_df_dummy['loan_income_ratio']= loan_df_dummy['loan_amnt']/loan_df_dummy['annual_inc']

# Extract Year & Month from Issue date
loan_df_dummy['month'],loan_df_dummy['year'] = loan_df_dummy['issue_d'].str.split('-', 1).str
loan_df_dummy[['issue_d','month','year']].head()

# Create Bins for range of Loan Amount
bins = [0, 5000, 10000, 15000, 20000, 25000,40000]
slot = ['0-5000', '5000-10000', '10000-15000', '15000-20000', '20000-25000','25000 and above']
loan_df_dummy['loan_amnt_range'] = pd.cut(loan_df_dummy['loan_amnt'], bins, labels=slot)

# Create Bins for range of Annual Income
bins = [0, 25000, 50000, 75000, 100000,1000000]
slot = ['0-25000', '25000-50000', '50000-75000', '75000-100000', '100000 and above']
loan_df_dummy['annual_inc_range'] = pd.cut(loan_df_dummy['annual_inc'], bins, labels=slot)

# Create Bins for range of Interest rates
if (loan_df_dummy['int_rate'].dtypes == str):
    loan_df_dummy['int_rate'] = pd.to_numeric(loan_df_dummy['int_rate'].str.strip('%'))
    
bins = [0, 7.5, 10, 12.5, 15,20]
slot = ['0-7.5', '7.5-10', '10-12.5', '12.5-15', '15 and above']
loan_df_dummy['int_rate_range'] = pd.cut(loan_df_dummy['int_rate'], bins, labels=slot)

```

```python
# Here, for annual Income outliers are present as max value is much far from mean value
# Removing outliers using Interquartile method

Q1 = loan_df_dummy['annual_inc'].quantile(0.25)
Q3 = loan_df_dummy['annual_inc'].quantile(0.75)
IQR = Q3 - Q1
loan_df_dummy['annual_inc'] = loan_df_dummy[~((loan_df_dummy['annual_inc'] < (Q1 - 1.5 * IQR)) |(loan_df_dummy['annual_inc'] > (Q3 + 1.5 * IQR)))]
```

```python
def crosstab(df,col):
    crosstab = pd.crosstab(df[col], df['loan_status'],margins=True)
    crosstab['Probability_Charged Off'] = round((crosstab['Charged Off']/crosstab['All']),3)
    crosstab = crosstab[0:-1]
    return crosstab
```

```python
  # Probability of charge off
def bivariate_prob(df,col,stacked= True):
    plotCrosstab = crosstab(df,col)
    
    linePlot = plotCrosstab[['Probability_Charged Off']]      
    barPlot =  plotCrosstab.iloc[:,0:2]
    ax = linePlot.plot(figsize=(20,8), marker='o',color = 'b')
    ax2 = barPlot.plot(kind='bar',ax = ax,rot=1,secondary_y=True,stacked=stacked)
    ax.set_title(df[col].name.title()+' vs Probability Charge Off',fontsize=20,weight="bold")
    ax.set_xlabel(df[col].name.title(),fontsize=14)
    ax.set_ylabel('Probability of Charged off',color = 'b',fontsize=14)
    ax2.set_ylabel('Number of Applicants',color = 'g',fontsize=14)
    plt.show()
```

```python
# Bivariate/Multivariate Analysis with Probability of Charge off

# Location vs Probability Charge Off
states = crosstab(loan_df_dummy,'addr_state')
states.head(20)
bivariate_prob(loan_df_dummy,col ='addr_state')

# There are multiple States/Provinces with high probability of charge
```
