### Data Cleaning

```python
# Drop duplicates
loan_df_dummy.drop_duplicates(inplace=True)
loan_df_dummy.shape
```

```python
# Remove columns having single unique value as they will not give value for our analysis
loan_df_dummy = loan_df_dummy[
    [column for column in list(loan_df_dummy) if loan_df_dummy[column].nunique() > 1]
]

# List of Columns & NA counts where NA values are more than 30%
NA_columns = loan_df_dummy.isnull().sum()
NA_columns = NA_columns[NA_columns.values >= (0.3 * len(loan_df_dummy))].index

# Drop NA Columns
loan_df_dummy.drop(labels=NA_columns, axis=1, inplace=True)
loan_df_dummy.shape

# Dropping unnecessary columns
loan_df_dummy.drop(
    [
        "id",
        "member_id",
        "url",
        "title",
        "delinq_2yrs",
        "revol_bal",
        "revol_util",
        "out_prncp",
        "out_prncp_inv",
        "total_pymnt",
        "total_pymnt_inv",
        "total_rec_int",
        "total_rec_late_fee",
        "total_rec_prncp",
        "recoveries",
        "collection_recovery_fee",
        "last_pymnt_d",
        "last_pymnt_amnt",
        "last_credit_pull_d",
    ],
    axis=1,
    inplace=True,
)
```

```python
# Removing rows with loan status as Current
loan_df_dummy = loan_df_dummy[
    loan_df_dummy.loan_status.isin(["Fully Paid", "Charged Off"])
]
loan_df_dummy.loan_status.unique()
```

```python
# Removing rows where pub_rec_bankruptcies is null
loan_df_dummy = loan_df_dummy[~loan_df_dummy.pub_rec_bankruptcies.isnull()]

# Removing rows where funded_amnt_inv is null
loan_df_dummy = loan_df_dummy[loan_df_dummy.funded_amnt_inv != 0]
```

```python
round(loan_df_dummy.isnull().sum() / loan_df_dummy.shape[0] * 100, 2)
```

### Derived Metrics

```python
# Extractig year and month from dates
loan_df_dummy["issue_month"], loan_df_dummy["issue_year"] = (
    loan_df_dummy["issue_d"].str.split("-", 1).str
)
```

```python
# Home Owenwrship: Replace Other and None in home_ownership with Unknown
loan_df_dummy["home_ownership"] = loan_df_dummy["home_ownership"].replace(
    ["OTHER", "NONE"], "UNKNOWN"
)
loan_df_dummy["home_ownership"].unique()
```

```python
# Employment Term : Remove 'n/a' value with 'self-employed'.
loan_df_dummy.emp_length.replace(["n/a", "NaN", None], "-1", inplace=True)

# Employee Title : Replace Blanks with "No Title Provided".
loan_df_dummy["emp_title"].fillna("No Title Provided", inplace=True)
```

```python
# Removing % sign from int_rate and revol_util acnd converting them to int
loan_df_dummy["int_rate"] = loan_df_dummy["int_rate"].str.replace("%", "").astype(float)
```

```python
# Loan amount to Annual Income percent
loan_df_dummy["loan_income_percent"] = round(
    100 * loan_df_dummy["loan_amnt"] / loan_df_dummy["annual_inc"], 2
)
loan_df_dummy[["loan_amnt", "annual_inc", "loan_income_percent"]].head()
```

```python
# Loan amount to Annual Income ratio
loan_df_dummy["installment_monthly_income_percent"] = round(
    100 * loan_df_dummy["installment"] / (loan_df_dummy["annual_inc"] / 12), 2
)
loan_df_dummy[
    ["installment", "annual_inc", "installment_monthly_income_percent"]
].head()
```

```python
# Convert emp_length to int

loan_df_dummy["emp_length"] = loan_df_dummy.emp_length.str.replace("years", "")
loan_df_dummy["emp_length"] = loan_df_dummy.emp_length.str.replace("year", "")
loan_df_dummy["emp_length"] = loan_df_dummy.emp_length.str.replace("<", "")
loan_df_dummy["emp_length"] = loan_df_dummy.emp_length.str.replace("+", "")
loan_df_dummy["emp_length"] = loan_df_dummy.emp_length.str.strip()
loan_df_dummy["emp_length"] = loan_df_dummy.emp_length.astype(int)
loan_df_dummy["emp_length"].unique()
```

```python
# Create Bins for range of Loan Amount
bins = [0, 5000, 10000, 15000, 20000, 25000, 40000]
slot = ["0-5000", "5000-10000", "10000-15000", "15000-20000", "20000-25000", "25000+"]
loan_df_dummy["loan_amnt_range"] = pd.cut(loan_df_dummy["loan_amnt"], bins, labels=slot)
loan_df_dummy[["loan_amnt", "loan_amnt_range"]]
```

```python
# Create Bins for range of Annual Income
bins = [0, 25000, 50000, 75000, 100000, 1000000]
slot = ["0-25000", "25000-50000", "50000-75000", "75000-100000", "100000+"]
loan_df_dummy["annual_inc_range"] = pd.cut(
    loan_df_dummy["annual_inc"], bins, labels=slot
)
loan_df_dummy[["annual_inc", "annual_inc_range"]]
```

```python
# Create Bins for range of Funded Amount by Investor
bins = [0, 5000, 10000, 15000, 20000, 25000, 40000]
slot = ["0-5000", "5000-10000", "10000-15000", "15000-20000", "20000-25000", "25000+"]
loan_df_dummy["funded_amnt_inv_range"] = pd.cut(
    loan_df_dummy["funded_amnt_inv"], bins, labels=slot
)
loan_df_dummy[["funded_amnt_inv", "funded_amnt_inv_range"]]
```

```python
# Create Bins for range of Annual Income
bins = [0, 2, 4, 6, 8, 10, 30]
slot = ["0-2", "2-4", "4-6", "6-8", "8-10", "10+"]
loan_df_dummy["installment_monthly_income_percent_range"] = pd.cut(
    loan_df_dummy["installment_monthly_income_percent"], bins, labels=slot
)
loan_df_dummy[
    ["installment_monthly_income_percent", "installment_monthly_income_percent_range"]
]
```

```python
# Create Bins for range of Annual Income
bins = [0, 2, 4, 6, 8,10, 30]
slot = ['0-2', '2-4', '4-6', '6-8', '8-10', '10+']
loan_df_dummy['installment_monthly_income_percent_range'] = pd.cut(loan_df_dummy['installment_monthly_income_percent'], bins, labels=slot)
loan_df_dummy[['installment_monthly_income_percent', 'installment_monthly_income_percent_range']]
```

```python
# Create Bins for range of Interest rates

bins = [0, 7.5, 10, 12.5, 15, 20]
slot = ["0-7.5", "7.5-10", "10-12.5", "12.5-15", "15 and above"]
loan_df_dummy["int_rate_range"] = pd.cut(loan_df_dummy["int_rate"], bins, labels=slot)
loan_df_dummy[["int_rate", "int_rate_range"]]
```

```python
# Function to plot countplot
def count_plot(x=None, y=None, hue=None, graph_height=10, graph_width=25):
    plt.figure(figsize=(graph_width, graph_height))
    ax = sns.countplot(x=x, y=y, hue=hue, data=loan_df_dummy)
    for p in ax.patches:
        height = 0 if str(p.get_height()) == "nan" else p.get_height()
        width = 0 if str(p.get_x()) == "nan" else p.get_x()
        if y is None:
            ax.annotate(
                "{:1.1f}%".format((height * 100) / float(len(loan_df_dummy))),
                (p.get_x() + 0.05, height + 20),
            )
        if x is None:
            ax.annotate(
                "{:1.1f}%".format((width * 100) / float(len(loan_df_dummy))),
                (p.get_height() + 0.05, width + 20),
            )
```

### Bivariate/Multivariate Analysis with Probability of Charge off

```python
# Printing graph for bivariate probability
def bivariate_prob(df, col, stacked=True):
    plotCrosstab = crosstab(df, col)
    linePlot = plotCrosstab[["Probability_Charged Off"]]
    barPlot = plotCrosstab.iloc[:, 0:2]
    ax = linePlot.plot(figsize=(15, 5), marker="o", color="b")
    ax2 = barPlot.plot(kind="bar", ax=ax, rot=1, secondary_y=True, stacked=stacked)
    ax2.set_ylabel("Number of Applicants", color="g", fontsize=14)
    plt.show()


# Compute a simple cross-tabulation of two (or more) factors. By default computes a frequency table
# of the factors unless an array of values and an aggregation function are passed
def crosstab(df, col):
    crosstab = pd.crosstab(df[col], df["loan_status"], margins=True)
    crosstab["Probability_Charged Off"] = round(
        (crosstab["Charged Off"] / crosstab["All"]), 3
    )
    crosstab = crosstab[0:-1]
    return crosstab
```

```python
# Location vs Probability Charge Off
crosstab(loan_df_dummy, "addr_state")
```

```python
bivariate_prob(df=loan_df_dummy, col="addr_state", stacked=False)
```

##### There are multiple States/Provinces with high probability of charge. Therefore, prediction cannot be made based on state

```python
# Annual Income vs Probability Charge Off
crosstab(loan_df_dummy, "annual_inc_range")
```

```python
bivariate_prob(df=loan_df_dummy, col="annual_inc_range", stacked=False)
```

##### Probability of charged off is inversely propotional to annual income

### Conclusion

#### Major variables to consider for loan prediction:
   
    1. Purpose of Loan => small business having more changes of charged off
    2. Ratio of Inatallment to Income is directly propotional to the chances of charged off
    3. Grade is inversely propotional to the chances of charged off
    4. Interest Rate is directly propotional to the chances of charged off
    5. Number of derogatory public records has high changes of being charged off for value 1 and 2
    6. Number of public record bankruptcies is directly propotional to the chances of charged off
    7. People who do not specify their home ownership have high changes of being charged off
    8. Annual salary is inversely propotional to the changes of being charged off

It is a good decision to approve loan to the applicants who have **high annual income** of **high grade** whose **installment generated will be less than 10% of monthly income** and has **no derogatory public records** and has **defined his home ownership**
