#### Cleaning Data

```python
# Getting missing values percentage
round(100 * (master_frame.isnull().sum() / len(master_frame.index)), 2)
```

```python
# Removing missing rows of raised_amount_usd
master_frame = master_frame[~np.isnan(master_frame["raised_amount_usd"])]
master_frame = master_frame[
    master_frame["funding_round_type"].isin(
        ("venture", "angel", "seed", "private_equity",)
    )
]

# Getting missing values percentage
print(round(100 * (master_frame.isnull().sum() / len(master_frame.index)), 2))
```

#### Means grouped by funding_round_type

```python
funding_round_mean = master_frame.groupby("funding_round_type")[
    "raised_amount_usd"
].mean()
print(funding_round_mean)
```

#### Considering that Spark Funds wants to invest between 5 to 15 million USD per  investment round, which investment type is the most suitable for them?

```python
FT = ""
for round_type in funding_round_mean.index:
    if 5000000 < funding_round_mean[round_type] < 15000000:
        FT = round_type
        print(round_type)
```

### Scrapping wiki page

#### Load data of wiki into bs4



```python
import requests
from bs4 import BeautifulSoup

URL = "https://en.wikipedia.org/wiki/List_of_territorial_entities_where_English_is_an_official_language"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
tables = soup.findAll("table")
english_speaking_countries = []
for table in tables[0:2]:
    rows = table.findAll("tr")
    for row in rows[1:]:
        columns = row.findAll("td")
        english_speaking_countries.append(
            {
                "country_name": columns[1].text.strip().split("[")[0],
                "country_code": columns[2].text.strip(),
            }
        )
english_speaking_countries = pd.DataFrame(english_speaking_countries)
english_speaking_countries
```

```python
del master_frame["_merge"]
master_frame = pd.merge(
    english_speaking_countries,
    master_frame,
    how="inner",
    left_on=["country_code"],
    right_on=["country_code"],
)
master_frame
```

### Checkpoint 6: Plots

#### A plot showing the fraction of total investments (globally) in angel, venture, seed, and private equity, and the average amount of investment in each funding type. This chart should make it clear that a certain funding type (FT) is best suited for Spark Funds.

```python
import matplotlib.pyplot as plt

funding_round_investment = (
    master_frame.groupby(["funding_round_type"])["raised_amount_usd"]
    .sum()
    .reset_index()
)
funding_round_investment["fraction"] = funding_round_investment[
    "raised_amount_usd"
].apply(
    lambda raised_amount_usd: raised_amount_usd
    / master_frame["raised_amount_usd"].sum()
)

funding_round_investment["mean"] = funding_round_investment["funding_round_type"].apply(
    lambda funding_round_type: master_frame[
        master_frame["funding_round_type"] == funding_round_type
    ]["raised_amount_usd"].mean()
)
del funding_round_investment["raised_amount_usd"]
funding_round_investment = funding_round_investment.set_index("funding_round_type")
axes = funding_round_investment.plot.bar(subplots=True, rot=0)
axes[1].legend(loc=2)
```
